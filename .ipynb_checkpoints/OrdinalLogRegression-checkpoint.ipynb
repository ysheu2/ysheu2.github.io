### What is ordinal logistic regression model and when should we use it?

In the previous example, I demonstrated that for data with repeated measures or a longitudinal design and a continuous outcome, a linear mixed-effect model is a suitable choice. This model assumes a normal distribution of errors and allows for the inclusion of both fixed and random effects. However, when your outcome variable is **<span style="font-size:14pt">ordinal**—such as low, medium, and high risk—the appropriate model is ordinal logistic regression. In this case, the cumulative odds of the ordinal categories in the model need to satisfy the proportional odds assumption. This assumption stipulates that the effect of predictor variables on the ordinal response remains constant across each category increase, ensuring the model's validity for ordered categorical outcomes.

In this example, we have two groups (Disease vs. Control), and we measured their FIB-4 score at baseline (prior to treatment), 6 months after treatment, and 12 months after treatment. We hypothesized that the trend of FIB-4 score over time will be different between the two groups (i.e. we expect an interaction between study group and timepoints). **However, this time, the FIB-4 score is converted into 3 fibrosis risk strata (low, intermediate, high) based on previous research.**


```python
import pandas as pd
import random
import numpy as np

# Set a seed for reproducibility
random.seed(108)

# Generating random data
study_ids = [f'STUDY_{i}' for i in range(1, 101)]
data = []

for study_id in study_ids:
    group = random.choice(['CONTROL', 'DISEASE'])
    age = random.randint(18, 65)
    sex = random.choice(['FEMALE', 'MALE'])
    
    for time_block in ['BASELINE', 'POST_6M', 'POST_12M']:
        fib4 = round(random.uniform(0, 5), 2)
        data.append([study_id, group, age, sex, time_block, fib4])

# Creating DataFrame
df = pd.DataFrame(data, columns=['STUDY_ID', 'GROUP', 'AGE', 'SEX', 'TIME_BLOCK', 'FIB4'])

# Define the categorization function
def categorize_fib4(value):
    if np.isnan(value):
        return np.nan
    elif value < 1.45:
        return 'low'
    elif 1.45 <= value <= 3.25:
        return 'medium'
    else:
        return 'high'

# Apply the categorization function to create the new column 'FIB4_category'
df['FIB4_category'] = df['FIB4'].apply(categorize_fib4)

# Convert the 'FIB4_category' column to a categorical variable with order
df['FIB4_category'] = pd.Categorical(df['FIB4_category'], categories=['low', 'medium', 'high'], ordered=True)

# Define a mapping from timepoints in category to numeric values
timepoint_mapping = {
    'BASELINE': 0,
    'POST_6M': 1,
    'POST_12M': 2
}

# Use the replace method to convert timepoints to numeric values
df['TIMEPOINT'] = df['TIME_BLOCK'].replace(timepoint_mapping)

# sort by MRN and Timepoint, then reset index to start sequentially with 0
df = df.sort_values(by=['STUDY_ID', 'TIMEPOINT']).reset_index(drop=True)

# Display the DataFrame with the new column
print(df.head())
```

### Violin graph showing each FIB4 Category (low, medium, high) distribution at different timepoints

Violin plots are especially useful when you want to compare the distribution of a variable across different categories. 

How to visualize a violin plot: higher height means higher density along the vertical axis, and wider width means higher density or concentration along the horizontal axis. 

As an example, at baseline (TIMEPOINT=0), most of the subjects in the control group fall within the "high" category, followed by "medium", then finally "low" category.  For disease group, most of the subjects fall within the "medium" and "high" categories, and a very noticable fewer subjects are in the "low" categroy.  When comparing post-6m (TIMEPOINT=1) to baseline (TIMEPOINT=0), you can see the change of category over time for the two groups.  In control group, the number of subjects in the low category decrease from baseline to post-6m. On the contrary, the number of subjects in the low category increase from baseline to post-6m for the disease group.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Create a violin plot
plt.figure(figsize=(10, 6))
sns.violinplot(x='TIMEPOINT', y='FIB4_category', hue='GROUP', data=df, split=True, inner='quart')
plt.title('Violin Plot of Categorical FIB4 Distribution Over Time for Control and Disease Groups')
plt.xlabel('Timepoint')
plt.ylabel('FIB4 Category')
plt.legend(title='Group', loc='upper right', bbox_to_anchor=(1, 1))
plt.savefig('Group_by_Timepoint_violin_graph.png')
plt.show()
```

## Ordinal Logistic Regression Model Specification: 

Unlike linear mixed-effects models, you don't include random intercepts or slopes to capture individual variability, because of the proportional odds assumption simplifies the model. The parameterization of OrderedModel from statsmodel requires that there is no constant in the model, neither explicit nor implicit. The constant is equivalent to shifting all thresholds and is therefore not separately identified.


```python
from statsmodels.miscmodels.ordinal_model import OrderedModel

# Define the formula with the interaction term (no need to add 0 to specify no constant)
formula = "FIB4_category ~ GROUP * TIMEPOINT + SEX + AGE"

# Fit the ordinal logistic regression model
ordinal_model = OrderedModel.from_formula(formula, data=df, distr='logit')
ordinal_result = ordinal_model.fit(method='bfgs')

# Print the results
print(ordinal_result.summary())
```

## Let's go through each variable in the model and interpret its coefficients:


1. **GROUP[T.DISEASE]:**
   - Coefficient: 0.0744
   - Interpretation: Holding other variables constant, individuals in the 'DISEASE' group are estimated to have a 0.0744 increase in the log-odds of moving to a higher category compared to the 'CONTROL' group. However, the p-value is 0.826, indicating that this effect is not statistically significant at a conventional significance level (e.g., 0.05).

2. **SEX[T.MALE]:**
   - Coefficient: 0.1338
   - Interpretation: Holding other variables constant, males are estimated to have a 0.1338 increase in the log-odds of moving to a higher category compared to females. The p-value is 0.533, suggesting that this effect is not statistically significant.

3. **TIMEPOINT:**
   - Coefficient: 0.0594
   - Interpretation: Holding other variables constant, a one-unit increase in the 'TIMEPOINT' variable is associated with a 0.0594 increase in the log-odds of moving to a higher category. The p-value is 0.738, indicating that this effect is not statistically significant.

4. **GROUP[T.DISEASE]:TIMEPOINT:**
   - Coefficient: -0.1842
   - Interpretation: This is the interaction effect between 'GROUP' and 'TIMEPOINT'. Holding other variables constant, the 'DISEASE' group's effect on the log-odds of moving to a higher category depends on the 'TIMEPOINT'. The negative coefficient suggests a decrease in the log-odds, but the p-value is 0.477, indicating that this interaction effect is not statistically significant.

5. **AGE:**
   - Coefficient: -0.0052
   - Interpretation: Holding other variables constant, a one-unit increase in age is associated with a -0.0052 decrease in the log-odds of moving to a higher category. The p-value is 0.503, suggesting that this effect is not statistically significant.

6. **low/medium:**
   - Coefficient: -0.9087
   - Interpretation: Individuals in the 'low/medium' category of the unspecified ordinal variable are estimated to have a -0.9087 decrease in the log-odds of moving to a higher category compared to the reference category. The p-value is 0.034, indicating that this effect is statistically significant.

7. **medium/high:**
   - Coefficient: 0.4152
   - Interpretation: Individuals in the 'medium/high' category of the unspecified ordinal variable are estimated to have a 0.4152 increase in the log-odds of moving to a higher category compared to the reference category. The p-value is 0.000, indicating that this effect is statistically significant.


## We will convert the coef in log-odds to odds ratio

```python
# Create a dictionary to store odds ratios
odds_ratios = {}

# Loop through the model results and calculate odds ratios
for param, coef in ordinal_result.params.items():
    # Skip the intercept terms
    if 'Intercept' in param:
        continue

    # Calculate odds ratio
    odds_ratio = np.exp(coef)

    # Store the odds ratio in the dictionary
    odds_ratios[param] = odds_ratio

# Print the results
for param, odds_ratio in odds_ratios.items():
    print(f"Odds Ratio for {param}: {odds_ratio:.3f}")
```
