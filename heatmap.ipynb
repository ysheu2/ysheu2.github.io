### A Simplified Way to Present Odds Ratios of Significant Predictors from Multiple Regression Models

In our analysis of GDMT prescribing patterns, I created five forest plots, each representing the significant predictors and their odds ratios for GDMT treatment in a specific medication class (ACEi/ARB/ARNI, Beta Blockers, MRA, SGLT2i, and Vasodilators). While forest plots are ideal for manuscripts to visually display predictors, having five of them is overwhelming for a scientific poster.

To address this, I condensed the information from all five forest plots into a single heatmap. In this heatmap, each cell represents the significance and effect size (odds ratios) of predictors across different models, offering a clear, compact visual summary!

```python
# load the toolboxes
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# load the excel sheet
df = pd.read_excel('heatmap_5classes.xlsx', sheet_name='treated_vs_not')

# take a look at the table
print(df.head(30))
```

```python
# We will use pivot_table to re-organize the table.
heatmap_data = df.pivot_table(values='Odds Ratio', index='Variable', columns='GDMT Class', aggfunc='mean')
```

```python
# Plot the heatmap
plt.figure(figsize=(10,6))
ax = sns.heatmap(heatmap_data, annot=True, cmap='coolwarm', center=1, linewidths=0.5, cbar_kws={'label':'Odds Ratio'})

# Set the aspect ratio
ax.set_aspect('auto')

# Set plot title and labels
plt.title('Odds Ratios Heatmap by Significant Variable and GDMT Class')
plt.xlabel('GDMT Medication Class')
plt.ylabel('Significant Variables')

plt.show()
```

This heatmap uses color coding (light to dark to indicate magnitude, and blue to red to show negative vs. positive correlations) to allow quick identification of variables that are consistently significant across different medication classes. This makes it easier for viewers to grasp overall trends at a glance.
